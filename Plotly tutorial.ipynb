We download district wise daily covid case data from covid india website. 

```python
districts_daily = requests.get('https://api.covid19india.org/data.json')
districts_daily.text
```

```python
districts_daily = districts_daily.text
districts_daily = json.loads(districts_daily)
```

```python
districts_daily.keys()
```

Data contains day wise case data for India. And statewise case data for states of India and testing data.

```python
districts_daily['cases_time_series'][0]
```

National data contains above given fields for each day. 

```python
len(districts_daily['statewise'])
```

We convert json data into pandas dataframe using below code

```python
totalactive = []
totalrecovered = []
totaldeceased = []
dailyactive = []
dailyrecovered = []
dailydeceased = []
date = []
for data in districts_daily['cases_time_series']:
    dailyactive.append(data['dailyconfirmed'])
    dailyrecovered.append(data['dailyrecovered'])
    dailydeceased.append(data['dailydeceased'])
    totalactive.append(data['totalconfirmed'])
    totalrecovered.append(data['totalrecovered'])
    totaldeceased.append(data['totaldeceased'])
    date.append(data['date'])
```

```python
df = pandas.DataFrame({'dailyactive': dailyactive, 'dailyrecovered': dailyrecovered, 'dailydeceased': dailydeceased,
                       'totalactive': totalactive, 'totaldeceased': totaldeceased, 'totalrecovered': totalrecovered,
                       'date': date})
```

The api contains data from January 2020 to August 2021. First five and last five rows shown below

```python
df.head()
```

```python
df.tail()
```

Json contains data in text format. So we parse it into appropriate format. 

```python
df['date'] = df['date'].str.strip()
df['date'] = pandas.to_datetime(df['date'], format = '%d %B %Y')
df['totalactive'] = df['totalactive'].astype(int)
df['totalrecovered'] = df['totalrecovered'].astype(int)
df['totaldeceased'] = df['totaldeceased'].astype(int)
df['dailyactive'] = df['dailyactive'].astype(int)
df['dailyrecovered'] = df['dailyrecovered'].astype(int)
df['dailydeceased'] = df['dailydeceased'].astype(int)
```

We have converted data into pandas dataframe.

Now I parse statewise corona virus data. 

```python
active = []
confirmed = []
deaths = []
deltaconfirmed = []
deltadeaths = []
deltarecovered = []
recovered = []
state = []
for data in districts_daily['statewise']:
    active.append(data['active'])
    confirmed.append(data['confirmed'])
    deaths.append(data['deaths'])
    deltaconfirmed.append(data['deltaconfirmed'])
    deltadeaths.append(data['deltadeaths'])
    deltarecovered.append(data['deltarecovered'])
    recovered.append(data['recovered'])
    state.append(data['state'])
```

```python
statedf = pandas.DataFrame({'active':active,
'confirmed':confirmed,
'deaths':deaths,
'deltaconfirmed':deltaconfirmed,
'deltadeaths':deltadeaths,
'deltarecovered':deltarecovered,
'recovered':recovered,
'state':state})
```

```python
statedf['active'] = statedf['active'].astype(int)
statedf['confirmed'] = statedf['confirmed'].astype(int)
statedf['deaths'] = statedf['deaths'].astype(int)
statedf['recovered'] = statedf['recovered'].astype(int)
statedf['deltaconfirmed'] = statedf['deltaconfirmed'].astype(int)
statedf['deltarecovered'] = statedf['deltarecovered'].astype(int)
statedf['deltadeaths'] = statedf['deltadeaths'].astype(int)
```

```python
statedf = statedf[statedf['state']!='Total']
```

Adding drop down menu for different chart types on same data

```python
fig = go.Figure()
fig.add_trace(go.Bar(x=df['date'], y=df['totalactive'], name='total active'))
fig.update_layout(title='Total Coronavirus Cases',                   
                  xaxis_title='Date',
                   yaxis_title='Active',showlegend= True)
fig.update_layout(
    updatemenus=[
        dict(active = 1,
            buttons=list([
                dict(
                    args=["type", "Bar"],
                    label="Bar Plot",
                    method="restyle"
                ),
                dict(
                    args=["type", "Scatter"],
                    label="Scatter",
                    method="restyle"
                )
            ])
        )
    ]
)

fig.show()
```

Adding dropdown menu for different data on same chart

```python
fig = go.Figure()
fig.add_trace(go.Bar(x=df['date'], y=df['totalactive'], name='total active'))
fig.add_trace(go.Bar(x=df['date'], y=df['totaldeceased'], name='total deceased'))
fig.add_trace(go.Bar(x=df['date'], y=df['dailyactive'], name='daily active'))
fig.add_trace(go.Bar(x=df['date'], y=df['dailydeceased'], name='daily deceased'))

fig.update_layout(
    updatemenus=[
        dict(
            active=0,
            buttons=list([
                dict(label="Total Active",
                     method="update",
                     args=[{"visible": [True, False, False, False]},
                           {"title": "Total Active"}]),
                dict(label="Total Deceased",
                     method="update",
                     args=[{"visible": [False, True, False, False]},
                           {"title": "Total Deceased"}]),
                dict(label="Daily Active",
                     method="update",
                     args=[{"visible": [False, False, True, False]},
                           {"title": "Yahoo Low"}]),
                dict(label="Daily Deceased",
                     method="update",
                     args=[{"visible": [False, False, False, True]},
                           {"title": "Daily Deceased"}])
            ]),
        )
    ])

# Set title
fig.update_layout(title_text="Coronavirus Cases")

fig.show()
```

Current Animation Limitations and Caveats
Animations are designed to work well when each row of input is present across all animation frames, and when categorical values mapped to symbol, color and facet are constant across frames. Animations may be misleading or inconsistent if these constraints are not met.
Although Plotly Express supports animation for many chart and map types, smooth inter-frame transitions are today only possible for scatter and bar
Plotly Express will not automatically compute the union of all x/y/color ranges, so these must be specified manually to avoid scale jumps across frames
